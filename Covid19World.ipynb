### Plotting using the covid19pyworld module.
Data source: https://github.com/CSSEGISandData/COVID-19.git

```python
import covid19pyworld as covid19
```

```python
covid19.plotMultiLine('confirmed',['US','Russia','India'])
```

```python
covid19.plotMultiLine('deaths',['US','Spain','Italy','Russia'])
```

```python
covid19.plotMultiLine('deaths',['Colombia','Mexico','Chile','Ecuador','Argentina'])
```

```python
covid19.plotReportedCasesBar('confirmed','US')
```

```python
covid19.plotReportedCasesBar('confirmed','Russia')
```

```python
covid19.plotReportedCasesLine('confirmed', 'Chile')
```

```python
covid19.plotReportedCasesLine('deaths', 'US')
```
